```python
import sqlite3
from datetime import datetime, timezone

conn = sqlite3.connect("logs/reviews.sqlite")
cur = conn.cursor()

cur.execute("""
INSERT INTO review_events
(timestamp, lemma, pos, result, translation, prompt_type)
VALUES (?, ?, ?, ?, ?, ?)
""", (
    datetime.now(timezone.utc).isoformat(),
    "lopen",
    "verb",
    1,
    "nl_to_en",
    "word"
))

conn.commit()
conn.close()
```
